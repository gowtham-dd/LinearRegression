```python
!pip install gradio
import pandas as pd
from sklearn import model_selection
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error, r2_score
import pickle
import gradio as gr
import matplotlib.pyplot as plt
import numpy as np

# Load the dataset
data = pd.read_csv('https://raw.githubusercontent.com/dataprofessor/data/refs/heads/master/delaney_solubility_with_descriptors.csv')

# Separate features and target
target = data['logS']
features = data.drop('logS', axis=1)

# Split data into training and testing sets
features_train, features_test, target_train, target_test = model_selection.train_test_split(
    features, target, test_size=0.2, random_state=42
)

# Train the linear regression model
lr = LinearRegression()
lr.fit(features_train, target_train)

# Save the model to a pickle file
filename = 'solubility_model.pkl'
pickle.dump(lr, open(filename, 'wb'))


def predict_solubility(MolWt, MolLogP, NumRotatableBonds, APSA):
    input_data = pd.DataFrame([[MolWt, MolLogP, NumRotatableBonds, APSA]],
                              columns=['MolWt', 'MolLogP', 'NumRotatableBonds', 'APSA'])
    prediction = lr.predict(input_data)[0]

    # Generate analysis chart
    plt.figure(figsize=(8, 6))
    plt.scatter(target_test, lr.predict(features_test), alpha=0.5)
    plt.plot(np.linspace(min(target_test), max(target_test), 100),
             np.linspace(min(target_test), max(target_test), 100), color='red')  # Diagonal line for reference
    plt.xlabel('Actual logS')
    plt.ylabel('Predicted logS')
    plt.title('Actual vs Predicted LogS (Test Data)')
    plt.scatter(target_test.iloc[0], lr.predict(features_test.iloc[[0]]), c='red', label='Current Input')
    plt.legend()

    plt.savefig("analysis_chart.png")  # Save the chart
    plt.close() # prevent multiple plots in the output
    return prediction, "analysis_chart.png"


# Create the Gradio interface
iface = gr.Interface(
    fn=predict_solubility,
    inputs=[
        gr.Number(label="Molecular Weight (MolWt)"),
        gr.Number(label="Molecular LogP (MolLogP)"),
        gr.Number(label="Number of Rotatable Bonds"),
        gr.Number(label="Apolar Surface Area (APSA)"),
    ],
    outputs=[
        gr.Number(label="Predicted logS"),
        gr.Image(label='Analysis Chart')
    ],
    title="Solubility Prediction",
    description="Predict the solubility of a molecule based on its descriptors.",
)

iface.launch(debug=True)
```

## RANDOM FOREST

```python
from sklearn.ensemble import RandomForestRegressor

rf = RandomForestRegressor(max_depth=2, random_state=42)
rf.fit(features_train, target_train)
```

```python
target_pred=rf.predict(features_train)
target_test_pred=rf.predict(features_test)
```

```python
from sklearn.metrics import mean_squared_error,r2_score

rf_train_mse=mean_squared_error(target_train,target_pred)
rf_train_r2=r2_score(target_train,target_pred)

rf_test_mse=mean_squared_error(target_test,target_test_pred)
rf_test_r2=r2_score(target_test,target_test_pred)
```

```python
print('LR MSE(Train)',lr_train_mse)
print('LR R2(Train)',lr_train_r2)
print('LR MSE(Test)',lr_test_mse)
print('LR R2(Test)',lr_test_r2)
```

```python
rf_results=pd.DataFrame(['Linear Regression',lr_train_mse,lr_train_r2,rf_test_mse,lr_test_r2]).transpose()
rf_results.columns=['Method','Training MSE','Training R2','Test MSE','Test R2']
rf_results
```
